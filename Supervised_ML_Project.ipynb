## Using SMOTE for equal sampling

```python
from imblearn.over_sampling import SMOTE

oversample = SMOTE()
X, y = oversample.fit_resample(X, y)

X = pd.DataFrame(X)
y = pd.DataFrame(y)

# checking the sizes of the sample data
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

## Logistic Regression

```python
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(solver='liblinear')
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
pred
```

```python
score=model.score(X_test,y_test)
score
```

```python
cm=metrics.confusion_matrix(y_test,pred,labels=[0,1])
df_cm=pd.DataFrame(cm, index=[i for i in ["0",'1']],
                   columns=[i for i in ['Predict 0',"Predict 1"]])
plt.figure(figsize=(7,5))
sns.heatmap(df_cm, annot=True)
```

```python
cf_matrix = metrics.confusion_matrix(y_test,pred)
print(cf_matrix)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score,precision_score,recall_score
print(accuracy_score(y_test, pred))
print(f1_score(y_test, pred, average=None))
print(precision_score(y_test, pred, average=None))
print(recall_score(y_test, pred, average=None))
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
print("Classification Report")
print(metrics.classification_report(y_test,pred,labels=[1,0]))
```

## Finding the best K

```python
from sklearn.model_selection import cross_val_score
scores_1=[]
for i in range(1,50):
  KNN_2=KNeighborsClassifier(n_neighbors=i)
  score_2=cross_val_score(KNN_2, X, y, cv=10)
  scores_1.append(score_2.mean())
```

```python
score_2
```

```python
scores_1
```

```python
plt.figure(figsize=(20,10))
plt.plot(range(1,50), scores_1, color='blue',linestyle='dashed', marker='o',markerfacecolor='red', markersize=10)
plt.title("Accuracy Rate Vs K_Values")
plt.xlabel('k')
plt.ylabel("Accuracy")
```
